```python
# Numpy
import numpy as np

# Pandas
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# Seaborn
import seaborn as sns

# RegEx
import re
```

```python
df = pd.read_json('https://raw.githubusercontent.com/ozlerhakan/mongodb-json-files/master/datasets/books.json',lines = True)
```

```python
for i in range(len(df.publishedDate)):
 df.publishedDate[i]=re.sub("{'\$.*: '|'}","",str(df.publishedDate[i]))
```

```python
for i in range(len(df.isbn)):
 df.isbn[i]=re.sub('[A-Za-z]*|\n+|-+|,+|:+','',str(df.isbn[i]))
```

```python
# Delete all the duplicate values
df.drop_duplicates('title',inplace=True)
```

```python
df.info()
```

```python
# Rearrange the values of the _id column
df['_id']= np.arange(1,428)
```

```python
df.info()
```

```python
df
```

```python
# Reindexing
df = df.reset_index(drop=True)
```

```python
# Replace the values in the pageCount column where the value == 0 with the mean of the
# mean (rounded off & dtype=int) of pageCount column elements

mean = round(np.mean(df.pageCount))
df['pageCount']=df['pageCount'].replace(0,mean)
```

```python
df.publishedDate.fillna("0",inplace=True)
```

```python
publishedYear1=[]
for i in range(len(df.publishedDate)):
 publishedYear1.append(re.sub("-.*","",df.publishedDate[i]))
```

```python
df['publishedYear']=publishedYear1
```
